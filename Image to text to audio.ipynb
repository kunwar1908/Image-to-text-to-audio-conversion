```python
import pytesseract
from PIL import Image
import gtts
```

```python
img = Image.open('/content/quote_advice_phrase_149482_3840x2160.jpg')
print(img)
```

```python
result = pytesseract.image_to_string(img)
print(result)
tts = gtts.gTTS(result)
tts.save("hello.mp3")
```
